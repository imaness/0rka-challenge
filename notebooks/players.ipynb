### Possible Birth Date Formats
- DD/MM/YY
- YYYY-MM-DD

```python
# Set new column for all formats

df_players_info.loc[df_players_info['Birthdate'].notnull() & df_players_info['Birthdate'].str.match(r'^\d{4}-\d{2}-\d{2}$'), "birthdate_format"] = "yyyymmdd"

df_players_info.loc[df_players_info['Birthdate'].notnull() & df_players_info['Birthdate'].str.match(r'^\d{2}\/\d{2}\/\d{2}$'), "birthdate_format"] = "ddmmyy"
```

```python
# Validate if there's more date formats
df_players_info[df_players_info['Birthdate'].notnull() & df_players_info['birthdate_format'].isnull()]
```

```python
# Check if DateOfDeath has same formats with birthdate

df_players_info.loc[df_players_info['Date_of_death'].notnull() & df_players_info['Date_of_death'].str.match(r'^\d{4}-\d{2}-\d{2}$'), "deathdate_format"] = "yyyymmdd"

df_players_info.loc[df_players_info['Date_of_death'].notnull() & df_players_info['Date_of_death'].str.match(r'^\d{2}\/\d{2}\/\d{2}$'), "deathdate_format"] = "ddmmyy"

df_players_info[df_players_info['Birthdate'].notnull() & df_players_info['birthdate_format'].isnull()]
```

```python
# Convert Dates to Type Dates
## Birthdate
df_players_info['Birthdate_datetime_type'] = pd.to_datetime(df_players_info['Birthdate'], format='%Y-%m-%d', errors='coerce')
df_players_info['Birthdate_datetime_type1'] = pd.to_datetime(df_players_info['Birthdate'], format='%d/%m/%y', errors='coerce', yearfirst=True)

### Merge two columns 
df_players_info['Birthdate_datetime_type'] = df_players_info['Birthdate_datetime_type'].combine_first(df_players_info['Birthdate_datetime_type1'])
df_players_info.drop('Birthdate_datetime_type1', axis=1, inplace=True)


## DeathDate
df_players_info['Date_of_death_datetime_type'] = pd.to_datetime(df_players_info['Date_of_death'], format='%Y-%m-%d', errors='coerce')
df_players_info['Date_of_death_datetime_type1'] = pd.to_datetime(df_players_info['Date_of_death'], format='%d/%m/%y', errors='coerce', yearfirst=True)

### Merge two columns 
df_players_info['Date_of_death_datetime_type'] = df_players_info['Date_of_death_datetime_type'].combine_first(df_players_info['Date_of_death_datetime_type1'])
df_players_info.drop('Date_of_death_datetime_type1', axis=1, inplace=True)


df_players_info
```

```python
# Check if there's is scenario where is age populated but not Birthdate
df_players_info[df_players_info['Age'].notnull() & df_players_info['Birthdate'].isnull()]
```

```python
# Correcting the Birthdate 

def birth_date_year_correction(birthdate, age):
    if birthdate is None:
        return None
    elif birthdate.year + age > datetime.now().year:
        # Assuming that 17th, 18th, 19th centuries are all in YYYY format
        return birthdate - pd.DateOffset(years=100)
    elif birthdate.year > datetime.now().year - 1:
        # Assuming that 17th, 18th, 19th centuries are all in YYYY format
        return birthdate - pd.DateOffset(years=100)
    else:
        return birthdate
    
df_players_info['Birthdate_corrected_datetime_type'] = df_players_info.apply(lambda row: birth_date_year_correction(row['Birthdate_datetime_type'], row['Age']), axis=1)
```

```python
df_players_info[df_players_info['Birthdate_corrected_datetime_type'] > datetime.now()]
```

```python
df_players_info['AgeDelta'] = df_players_info['Date_of_death_datetime_type'].dt.year - df_players_info['Birthdate_corrected_datetime_type'].dt.year
df_players_info['AgeDelta'].fillna(0, inplace=True)
```

```python
df_players_info[df_players_info['AgeDelta'] < 0]
```

```python
# Validate if there are still incorrect Death Dates
df_players_info[(df_players_info['Date_of_death_datetime_type'] - df_players_info['Birthdate_corrected_datetime_type']).dt.days / 365 - df_players_info['Age'] > 2]
```

```python
df_players_info['AgeDigit'] = df_players_info['Age'].fillna(0)

# Correcting the Death date 
def death_date_year_correction(death_date, birthdate, age, id):
    if age == 0:
        return None
    elif (death_date.year - (birthdate.year + age)) > 0:
        # Assuming that 17th, 18th, 19th centuries are all in YYYY format
        return death_date - pd.DateOffset(years=100)
    else:
        return death_date
df_players_info['Date_of_death_corrected_datetime_type'] = df_players_info.apply(lambda row: death_date_year_correction(row['Date_of_death_datetime_type'], row['Birthdate_corrected_datetime_type'], row['AgeDigit'], row['ID']), axis=1)
```

```python
# Validate if there are still incorrect Death Dates

df_players_info['AgeCorrected'] = np.floor((df_players_info['Date_of_death_corrected_datetime_type'] - df_players_info['Birthdate_corrected_datetime_type']).dt.days  / 365)
```

```python
df_players_info[(np.abs(df_players_info['AgeCorrected'] - df_players_info['Age']) > 1) & (df_players_info['Date_of_death'].notnull())]
```
